```python
#Loading the data set
hotel = pd.read_csv(r"/content/hotel_bookings.csv")
```

```python
hotel['reservation_status_date'] = pd.to_datetime(hotel['reservation_status_date'], format="%d-%m-%Y")
```

```python
#removing the null values
hotel.isnull().sum()
```

```python
hotel.drop(['company','agent'], axis = 1, inplace = True)
```

```python
hotel.dropna(inplace = True)
```

```python
hotel.isnull().sum()
```

```python
hotel.describe()
```

```python
hotel = hotel[hotel['adr']<5000]
```

```python
#Data analysis and Visualizations
```

```python
#gives reservation status count
cancelled_perc = hotel['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],hotel['is_canceled'].value_counts(), edgecolor = 'k',width = 0.7)
```

```python
#Resevation status in different hotels
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = hotel, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
plt.title('Resevation status in different hotels', size = 20)
plt.xlabel('hotels')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel = hotel[hotel['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = hotel[hotel['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#Have you ever wondered when the best time of year to book a hotel room is?
hotel['month'] = hotel['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = hotel, palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number  of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#the optimal length of stay in order to get the best daily rate?
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
#What if you wanted to predict whether or not a hotel was likely to receive a disproportionately high number of special requests?
plt.figure(figsize=(12, 12), dpi=80)

plt.subplot(4, 2, 4)
plt.title('Total of Special Requests')
sns.countplot(x='total_of_special_requests', data=hotel, palette='coolwarm')

plt.tight_layout()
plt.show()
```
